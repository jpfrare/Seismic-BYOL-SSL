```python
import numpy as np
import pandas as pd
```

```python
file_path = "../reports/evaluation_results.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
grouped = df.groupby(["downstream", "cap", "backbone"])['iou'].agg(['mean', 'std']).reset_index()
grouped
```

```python
grouped['backbone'].unique()
```

```python
grouped.replace({
    "seam_ai_norm": "seam_ai_N",
    "IMAGENET": "imagenet",
    "COCO": "coco",
    "f3_norm": "f3_N"
}, inplace=True)
```

```python
grouped.rename(columns={"backbone": "pretrain_data", "downstream": "train_data"}, inplace=True)
grouped['cap'] = (grouped['cap'] * 100).astype(float)
grouped['cap'].unique()
```

```python
grouped.to_csv("grouped.csv", index=False)
```
